```python
tmdb_movie = Movie()
def get_genre(x): #pass in the title of the movies
    genres = []
    result = tmdb_movie.search(x) #the title will be searched in the tmdb_movie
    movie_id = result[0].id #we will match the "id" with the "title"
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key)) #we will get the result from the IMDb data
    data_json = response.json() #we will then convert it to a json file
    if data_json['genres']: #in the json file we will only need to extract the "genre"
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name']) #we will then add the "genre" to the empty genre list we created above
        return genre_str.join(genres)
    else:
        np.NaN # we will return the results but if we don't find anything we will consider it as a missing value
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

### **Extract Directors**

```python
def get_director(x):
    if " (director)" in x: #we are getting "director"
        return x.split(" (director)")[0]
    elif " (directors)" in x: #we are getting "directors" with "s"
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0] #we are getting "directors/screenplay"
```

```python

#let's apply the above function on the "Cast and crew" column
data_2018['director_name'] = data_2018['Cast and crew'].map(lambda x: get_director(x))
```

### **Actor 1**

```python
data_2018['Cast and crew'][0].split("screenplay")[-1].split(",")[0]
```

```python
# Let define fucntion for extracting actor1
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
#let's apply the above function on the "Cast and crew" column
data_2018['actor_1_name'] = data_2018['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
data_2018.head()
```

### **Actor 2**

```python
len(data_2018['Cast and crew'][3].split("screenplay);")[-1].split(","))
```

```python
# Let define fucntion for extracting actor2
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
#let's apply the above function on the "Cast and crew" column
data_2018['actor_2_name'] = data_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

### **Actor 3**

```python
# Let define fucntion for extracting actor3
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
#let's apply the above function on the "Cast and crew" column
data_2018['actor_3_name'] = data_2018['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
#dataset for 2018 movie final
new_data18 = data_2018.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

## **Extracting features of 2020 movies from Wikipedia**

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2020"

source = urllib.request.urlopen(link).read()
soup = bs.BeautifulSoup(source,'lxml')
```

```python
tables = soup.find_all('table',class_='wikitable sortable')
```

```python
df1 = pd.read_html(str(tables[0]))[0]
df2 = pd.read_html(str(tables[1]))[0]
df3 = pd.read_html(str(tables[2]))[0]
df4 = pd.read_html(str(tables[3]).replace("'1\"\'",'"1"'))[0] # avoided "ValueError: invalid literal for int() with base 10: '1"'
```

```python
data = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
data
```

```python
data.shape
```

```python
data.columns
```

```python
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
        return np.NaN
    else:
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.NaN
```

```python
data['genres'] = data['Title'].map(lambda x: get_genre(str(x)))
```

```python
data
```

```python
#Renaming the columns
data_2020 = data_2020.rename(columns={'Title':'movie_title'})
```

```python
#Extracting the columns which are required for training
new_data20 = data_2020.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```
